```python
from langchain_community.llms import Cohere
from langchain_community.embeddings import CohereEmbeddings

cohere_api_key = "API KEY"

# Initialize Cohere LLM
llm = Cohere(cohere_api_key=cohere_api_key)

# ‚úÖ Fix applied: pass `user_agent="langchain"` to avoid KeyError
embedding = CohereEmbeddings(
    cohere_api_key=cohere_api_key,
    user_agent="langchain"
)

print("‚úÖ Cohere LLM and Embeddings initialized successfully.")
```

```python
# ‚úÖ Import cassio
import cassio

# ‚úÖ Initialize Astra DB connection
cassio.init(
    token="YOUR TOKEN",
    database_id="YOUR ID",
    keyspace="YOUR KEY SPACE"
)
```

```python
from langchain_community.vectorstores import Cassandra
import cassio

# ‚úÖ Use resolve_session() and resolve_keyspace()
astra_vector_store = Cassandra(
    embedding=embedding,
    table_name="qa_mini_demo",
    session=cassio.config.resolve_session(),
    keyspace=cassio.config.resolve_keyspace()
)
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter=CharacterTextSplitter(
    separator="\n",
    chunk_size=800,
    chunk_overlap=200,
    length_function=len,
)
texts=text_splitter.split_text(raw_text)
```

```python
texts[:50]
```

```python
from langchain.indexes.vectorstore import VectorStoreIndexWrapper

# After inserting texts into the vector DB
astra_vector_store.add_texts(texts[:50])
print("‚úÖ Inserted %i documents." % len(texts[:50]))

# Wrap it for question answering
astra_vector_index = VectorStoreIndexWrapper(vectorstore=astra_vector_store)
```

```python
first_question = True
while True:
    if first_question:
        query_text = input("\nEnter your question (or type 'quit' to exit): ").strip()
    else:
        query_text = input("\nWhat's your next question (or type 'quit' to exit): ").strip()

    if query_text.lower() == "quit":
        break

    if query_text == "":
        continue

    first_question = False

    print("\nQUESTION: \"%s\"" % query_text)
    answer = astra_vector_index.query(query_text, llm=llm).strip()
    print("ANSWER: \"%s\"\n" % answer)

    print("FIRST DOCUMENTS BY RELEVANCE:")
    for doc, score in astra_vector_store.similarity_search_with_score(query_text, k=4):
        print("    [%0.4f] \"%s ...\"" % (score, doc.page_content[:84]))
```
